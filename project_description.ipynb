# Video URL & Description

**URL:** [Starbucks Bear Cup Controversy](https://www.tiktok.com/t/ZTMtmqpYy/)

**Description:** A woman discusses what she views as the embarassing culture of overconsumption in America, using the recent Starbucks Bear Cup craze as a key example supporting her claim. For context, Starbucks released a limited-edition cup in the shape of a teddy bear, and excited customers were camping outside Starbucks shops as early as the night before to try to retrieve one. As well, customers have been seen on video fighting each other and with Starbucks workers over these cups. <br>
The comments *generally* either agree with the original poster points on calling waiting for Starbucks cups a worthless endeavor, while others say that the original poster shouldn't care what other people do with their lives.

# Research Question

To what extent do commentators use humor to criticize consumerist behavior?

# Concepts & How They Changed

Our original two concepts were:
> **Concept1** = "Code 1 if the comment agrees with the video--waiting for the starbucks bear cup is embarrassing. Code 0 if the comment doesn't agree with the video."
>
> **Concept2** = "Code 1 if the comment disagrees with the video--individual choice should be respected or whether they wait doesn't matter. Code 0 if the comment doesn't disagree with the video."

As we were working through our initial coding round, we ended up further refining our research question as we became more familiar with the comment database available to us. Initially, it was "to what extent do the commentators agree with the video. However, as we noticed the variation in the amount of humor being used in the comments, we refined our research question into, "to what extent do commentators use humor to criticize consumerist behavior?" This change led to adjustments to our concepts. 

Our refined final concepts are:
> **concept1** = "Code 1 if the comment incorporates humor--use of laughing emojis, indirect humorous commentary, comparisons, sarcasm or irony, jokes, etc. Code 0 if the comment does not include humor or any of these elements."
>
> **concept2** = "Code 1 if the comment directly criticizes consumerist behavior (e.g. agrees with the video, finds the situation embarassing, the commenter does not identify with the behavior, etc). Code 0 if the comment does not criticize consumerist behavior or does not specify one position."

As we coded, we realized that there were a few edge cases that our concepts didn't quite address. For example, comments that are more subtle or indirect in agreeing with the video (e.g. laughing, sharing analogies to the situation which may be seen as humorous, etc). We addressed the comments in different ways, and so to address this, we agreed that any subtlety that show agreement would be coded as 1 for concept 2. We included these specific cases in concept 2. Also, for the comments that both agreed and disagreed (e.g. Yes agree but you're also sad for caring so much), we decided to code them as 0 for concept 2.

# Final LLM Prompt & Inter-Rater Reliability Scores

**This is the final LLM prompt with the following inter-rater reliability scores:**

You are a research assistant coding TikTok comments for a project examining how humor is used in criticisms of overconsumerism. You will be given individual comments responding to a TikTok video that mocks overconsumerism and, specifically, the eagerness of people trying to obtain the Starbucks bear cup. Follow these rules EXACTLY. If criteria for “1” are not clearly met, default to “0.” 

Follow this codebook to code for concepts 1 and 2:
- concept1:
    - Code 1 only if at least one of the following elements are present:
        - if the comment incorporates humor--use of laughing or crying emojis
        - uses text laughs like “lol”, “lmao”, “lmfao”, “haha”, “i’m dead”
        - indirect humorous commentary
        - humorous comparisons
        - sarcasm or irony
        - rhetorical questions
        - jokes, profanity
        - slang in a humorous way (e.g. "girl")
    - Code 0 if the comment does not include the elements at least one of the elements of humor noted above OR if it includes at least one of the following: 
        - straightforward criticism without comedic cues
        - purely descriptive or factual
        - expressing sincere frustration or disapproval with no playful or humorous framing
        - simply states "It’s crazy!!!".
- concept2:
    - Code 1 only if at least one of the following elements are present:
        - states "thank you" or shows gratitude or "yes!"
        - "It’s crazy!!!"
        - if the comment directly criticizes consumerist behavior
        - states that they agree
        - Calls the behavior embarrassing, odd, absurd, crazy, wasteful, ridiculous
        - the commenter does not identify with the behavior
        - commentor states “couldn’t be me,” “I wouldn’t,” “you couldn’t pay me,” “not worth it,”
        - Linking behavior to debt, waste, clutter, landfill harm, irresponsible spending (e.g. "you ppl didn't play when u were young. too much individualism kids don't play with each other. still fascinated by teddy bears which should be left to kids. all these cups will end up at Goodwill that's why car garages are filled with junk")
    - Code 0 if the comment does not follow at least one of the criteria above OR if it includes at least one of the following:
        - states two opposing opinions about the topic or situation
        - brings up the starbucks strike or boycott (e.g who's buying starbucks?)
        - states, "Why is anyone still going to Starbucks is the real question!"
        - talks about resellers.
        - talks about china
        - talks about the item itself, not the behavior
        - shows sympathy for workers
        - discussion of marketing
        - states "It's embarrassing to care that much about what people spend their own money on like why are you so mad? however, it is also just as embarrassing to fight over said cup or to buy them up for themselves."
        - shares unrelated comparisons to the topic of consumerism

**Krippendorff’s Alpha between Human Coders:**

**Krippendorff’s Alpha between Human and AI:**
- **Concept 1: 0.735**

# Reflections on Improving Scores

It was a very long process to improve the inter-rater reliability scores between the LLM and our human coding. First, we realized quickly that the LLM was not the most adept at identifying humor and that it required much more guidance in identifying it. Due to this, we gave much more specific examples and instructions that were based on the actual comments and used its wording and incorporated it into the codebook in the LLM prompting. This would ensure that the LLM would be able to have more information on what specifically they could or could not include--thus, we had to ultimately moreso hardcode some items. In addition to this, another big change we made is how we structured the codebook as a whole. Originally, we only formatted the codebook as simply, "code 1 for XXX, and code 0 for XXX." However, we found that revising the prompt to be, "Code 1 if it includes at least one of the following elements: XXX. Code 0 if it does not have at least one of the elements above OR if it includes the following elements: XXX." We believe that this formatting was more efficient for our coding process because it introduced structure into how the LLM would follow a decision tree as it was coding between 1 and 0 for each concept. This added structure allowed for improvement to our inter-rater reliability scores. Though there were still some disagreements we couldn't change even with the modified prompting specifying how they should code a specific comment, overall, this modification to the prompt improved the inter-rater reliability scores to a substantial degree.

## Written Report

Humor clearly plays a big role in how people engage. Funny comments got the most likes by far, while opinion-heavy ones *generally* got much less attention. That makes sense in a debate-heavy comment section, as so many comments are flooding in opinions that it makes sense that most of them don't get attention, since they're the majority anyway. We see the same pattern with replies too: humorous comments spark more interaction, while strongly opinionated ones don’t take off as much. <br> <br>
Neutral, non-humorous comments actually got the most visibility from TikTok’s algorithm, while supportive opinionated comments were shown the least. Demographically, commenters look pretty similar across all concepts—mostly ages 25–30, with spillover into 18–24 and 30–34, so overall a Gen Z + young millennial crowd. <br> <br>
Linguistically, the more opinionated comments used more complex wording, likely because people put more thought into expressing a stance. Meanwhile, humorous comments were the shortest and punchiest, which fits their style.
